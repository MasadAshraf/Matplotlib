```python
import matplotlib.pyplot as plt
from PIL import Image


# load image with operating system image viwer application
img = Image.open('content/image.JPG')
plt.imshow(img)
```

# Converting image into Grey Scale

```python
#Image.convert() Returns a converted copy of this image.
img2 = img.convert('L')
plt.imshow(img2, cmap=plt.get_cmap('gray')) 
```
